```python
import matplotlib.pyplot as plt
import numpy as np 
import cv2
```

```python
raw_images = []

for i in range(1,3):
    img_3d = None
    img_3d = np.load("img_data_full/{}.npy".format(i))
    raw_images.append(img_3d)
    
print(len(raw_images))
```

```python
img_3d = raw_images[1]
img_3d.shape
```

```python
plt.imshow(img_3d[:,:,100])
```

```python
img_3d.shape
```

```python
w, h = 128, 128

def rs_img(img):
    flatten = [cv2.resize(img[:,:,i], (w, h), interpolation=cv2.INTER_CUBIC) for i in range(img.shape[-1])]
    img = np.array(np.dstack(flatten)) 
    return img
```

```python
img_rs = rs_img(img_3d)
img_rs.shape
```

```python
plt.imshow(img_rs[:,:,120])
```

```python
np.save("v.npy", img_rs)
```
